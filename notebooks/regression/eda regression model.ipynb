```python
# Mapping van maandnamen naar maandnummers
# Deze mapping is nodig om de maandnamen in de CBS dataset om te zetten naar maandnummers, zodat we een correcte Datetime variabele kunnen maken.
maanden_dict = {
    'januari': '01', 'februari': '02', 'maart': '03', 'april': '04',
    'mei': '05', 'juni': '06', 'juli': '07', 'augustus': '08',
    'september': '09', 'oktober': '10', 'november': '11', 'december': '12'
}
```

Inladen dataset CBS aantal autoverkopen en mergen met Bovemij dataset

```python
# Laden dataset CBS autoverkopen totaal per maand
df_cbs_autoverkopen = pd.read_csv('../../data/external/Verkochte_wegvoertuigen_09032026_090156.csv', sep=';')
```

```python
df_cbs_autoverkopen.head()
```

```python
# Data cleaning en transformatie van de CBS dataset

# 1. Verwijder het sterretje (*) van voorlopige cijfers en spaties
df_cbs_autoverkopen['Perioden'] = df_cbs_autoverkopen['Perioden'].str.replace('*', '', regex=False).str.strip()

# 2. Splits '2021 januari' in twee losse kolommen: '2021' en 'januari'
df_cbs_autoverkopen[['Jaar', 'Maand_Naam']] = df_cbs_autoverkopen['Perioden'].str.split(' ', expand=True)

# 3. Koppel het maandnummer eraan en maak een correcte Datetime variabele (YYYY-MM-01)
df_cbs_autoverkopen['Maand_Num'] = df_cbs_autoverkopen['Maand_Naam'].map(maanden_dict)
df_cbs_autoverkopen['Datum'] = pd.to_datetime(df_cbs_autoverkopen['Jaar'] + '-' + df_cbs_autoverkopen['Maand_Num'] + '-01')

# 4. We plakken Jaar en Maand_Num aan elkaar en maken er een integer van voor de merge
df_cbs_autoverkopen['Jaar_Maand'] = (df_cbs_autoverkopen['Jaar'] + df_cbs_autoverkopen['Maand_Num']).astype(int)
```

```python
df_cbs_autoverkopen.head()
```

```python
# Filter de data vanaf Jaarmaand 201501
df_cbs_autoverkopen = df_cbs_autoverkopen[df_cbs_autoverkopen['Jaar_Maand'] >= 201501].reset_index(drop=True)
```

```python
# Get unique values for Product and Label_Naam
unique_voertuigen = df_cbs_autoverkopen['Wegvoertuigen'].unique()
unique_brandstof = df_cbs_autoverkopen['Brandstofsoort voertuig'].unique()

# Count unique values
num_voertuigen = len(unique_voertuigen)
num_brandstof = len(unique_brandstof)

print(f"Unique Vehicles ({num_voertuigen}):")
print(unique_voertuigen)
print(f"\nUnique Fuel Types ({num_brandstof}):")
print(unique_brandstof)
```

```python
df_cbs_autoverkopen.info()
```

```python
df_cbs_autoverkopen.describe()
```

```python
# We plakken de CBS data naast de Bovemij data. 

# 1. Voorbereiden van de CBS data
cbs_col_name = 'Totaal verkochte wegvoertuigen (aantal)'

# We maken een kopie en zorgen dat 'Jaar_Maand' beschikbaar is als kolom
# (Ik ga er vanuit dat df_cbs_autoverkopen de kolom 'Jaar_Maand' al heeft, 
# anders moet je de 'Perioden' kolom eerst omzetten zoals we eerder bespraken)
cbs_subset = df_cbs_autoverkopen[['Jaar_Maand', cbs_col_name]].copy()
cbs_subset = cbs_subset.rename(columns={cbs_col_name: 'CBS_Autoverkopen_Totaal'})

# 2. Mergen op de kolom 'Jaar_Maand'
# We gebruiken 'left' join om alle rijen van je Bovemij data (df) te behouden
df_merged_1 = pd.merge(df, cbs_subset, on='Jaar_Maand', how='left')
```

Inladen CBS CPI Verzekeringen ivm Vervoer dataset inladen en mergen met Bovemij dataset.

```python
# Laden dataset CBS 
df_cbs_cpi_verzekeringen = pd.read_csv('../../data/external/CPI__prijsindex_2015_100_09032026_104238_125400_verzekeringen_ivm_vervoer.csv', sep=';')


# Data cleaning en transformatie van de CBS dataset
# 1. Verwijder het sterretje (*) van voorlopige cijfers en spaties
df_cbs_cpi_verzekeringen['Perioden'] = df_cbs_cpi_verzekeringen['Perioden'].str.replace('*', '', regex=False).str.strip()

# 2. Splits '2021 januari' in twee losse kolommen: '2021' en 'januari'
df_cbs_cpi_verzekeringen[['Jaar', 'Maand_Naam']] = df_cbs_cpi_verzekeringen['Perioden'].str.split(' ', expand=True)

# 3. Koppel het maandnummer eraan en maak een correcte Datetime variabele (YYYY-MM-01)
df_cbs_cpi_verzekeringen['Maand_Num'] = df_cbs_cpi_verzekeringen['Maand_Naam'].map(maanden_dict)
df_cbs_cpi_verzekeringen['Datum'] = pd.to_datetime(df_cbs_cpi_verzekeringen['Jaar'] + '-' + df_cbs_cpi_verzekeringen['Maand_Num'] + '-01')


# Filter de data vanaf Datum 2015-01-01 en reset de index voor een correcte joining later
df_cbs_cpi_verzekeringen = df_cbs_cpi_verzekeringen[df_cbs_cpi_verzekeringen['Datum'] >= '2015-01-01'].reset_index(drop=True)   
```
